```python
import cv2  #pip install opencv-python
import os   #pip install os-sys

image_dir = '/home/arturmarcon/code/ChilleeX/NoLateX/initial_test_data/batch_1_salmple100' #Dir containing images
archives = os.listdir(image_dir) #listing all dir files

exit_dir = '/home/arturmarcon/code/ChilleeX/NoLateX/Binarization/binarization_img_otsu'
if not os.path.exists(exit_dir):
    os.makedirs(exit_dir)


def binarization_image(image): #Function to binarizate one image
    _, binarization_img = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) #applied the binarization
    return binarization_img

for archive in archives:
    if archive.endswith('.jpg'):
        image_path = os.path.join(image_dir, archive) #load image
        image = cv2.imread(image_path, 0)
        binarization_img = binarization_image(image) #binarize image

        exit_path = os.path.join(exit_dir, f'binarizada_{archive[:-4]}.jpg') #save binarized image
        cv2.imwrite(exit_path, binarization_img)

        if os.path.exists(exit_path):
            print(f'{archive} binarized image_{archive}')
        else:
            print(f'Erro: didnt save {archive}.')
```
